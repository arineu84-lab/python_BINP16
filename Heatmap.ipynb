Script name: 
    PlotDistMatrices.py

Description: 
    Based on a genetic distance matrix (calculated by another student), this script should plot a heatmap. Here, the heatmap will be displayed clustered with labels for "Romanov" and "non-Romanov", the package used for this will be "seaborn". Based on this heatmap clustering, a potential relation between one "non-Romanov" profile with the Romanov family can be hypothesised.

User defined functions:

Procedure: 
    1. Import pandas, matplotlib and seaborn 
    2. Define "Romanov" cluster
    3. Assign the label and set colours
    4. Create heatmap with certain parameters

Input: 
    input_file (This is the output_file of student 2)

Output: 
    heatmap_mtDNA/Y.png

Usage: 
    python Heatmap.py input_file output_file

```python
# pip install seaborn

import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
# Load the genetic distance matrix
genetic_matrix = pd.read_csv('genetic_distance_matrix.csv', sep='\t', index_col=0)

# Define Romanov identifiers
romanov_cluster = [
    'Princess Irene', 'Prince Fred', 'Nicolas II Romanov', 'Alexandra Romanov',
    'Olga Romanov', 'Tatiana Romanov', 'Maria Romanov', 'Alexei Romanov',
    'Suspected body of Anastasia Romanov']

# Assign labels
labels = ['Romanov' if name in romanov_cluster else 'non-Romanov' for name in genetic_matrix.index]

# Map labels to colours: purple for Romanov, teal for non-Romanov
color_code = {'Romanov': 'purple', 'non-Romanov': '#d0f0ff'}
row_colors = pd.Series(labels, index=genetic_matrix.index).map(color_code)

# Create clustered heatmap with row colors
sb.set(font_scale=1.1)
sb.clustermap(genetic_matrix, cmap='viridis', linewidths=0.5, figsize=(12, 10), row_colors=row_colors)

# Show the plot
plt.show()
```
